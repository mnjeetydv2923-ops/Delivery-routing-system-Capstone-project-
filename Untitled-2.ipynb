```python
def route_cost(path, graph):
    # Base case
    if len(path) == 1:
        return 0

    # Check if edge exists
    if path[1] not in graph[path[0]]:
        return float('inf')  # no connection

    return graph[path[0]][path[1]] + route_cost(path[1:], graph)


# Graph definition
graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

# Path
path = ['A', 'B', 'C', 'D']

# Calculate cost
cost = route_cost(path, graph)

# Output
print("Path:", " -> ".join(path))
print("Route Cost =", cost)
```

```python
# Sample data (make sure this exists)
capacity = 5
parcels = [
    {"id": 1, "weight": 2, "value": 40, "deadline": 2},
    {"id": 2, "weight": 3, "value": 50, "deadline": 3},
    {"id": 3, "weight": 5, "value": 100, "deadline": 1}
]

# Greedy: sort by value/weight ratio
parcels_sorted = sorted(parcels, key=lambda x: x['value']/x['weight'], reverse=True)

selected = []
total_weight = 0
total_value = 0

for p in parcels_sorted:
    if total_weight + p['weight'] <= capacity:
        selected.append(p)
        total_weight += p['weight']
        total_value += p['value']


selected_sorted = sorted(selected, key=lambda x: x['deadline'])

time = 0
feasible = True

for p in selected_sorted:
    time += 1  # each delivery takes 1 unit time
    if time > p['deadline']:
        feasible = False
        break



print("Selected Parcels:", [p['id'] for p in selected])
print("Total Weight:", total_weight)
print("Total Value:", total_value)
print("Time Window Feasible:", feasible)
```

```python
import heapq   # ✅ correct import

def dijkstra(graph, start):
    pq = [(0, start)]
    dist = {node: float('inf') for node in graph}
    dist[start] = 0

    while pq:
        cost, node = heapq.heappop(pq)

        if cost > dist[node]:
            continue

        for neighbor, weight in graph[node].items():
            new_cost = cost + weight

            if new_cost < dist[neighbor]:
                dist[neighbor] = new_cost
                heapq.heappush(pq, (new_cost, neighbor))

    return dist


# Graph definition (REQUIRED)
graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

# Run algorithm
shortest_paths = dijkstra(graph, 'A')

print("Shortest Paths from A:", shortest_paths)
```

```python
from itertools import permutations

def tsp(graph, start):
    nodes = list(graph.keys())
    nodes.remove(start)

    min_cost = float('inf')
    best_route = None

    for perm in permutations(nodes):
        current = start
        cost = 0

        for nxt in perm:
            cost += graph[current][nxt]
            current = nxt

        cost += graph[current][start]

        if cost < min_cost:
            min_cost = cost
            best_route = (start,) + perm + (start,)

    return best_route, min_cost


# Example graph (you can modify this)
graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

route, min_cost = tsp(graph, 'A')

print("Optimal Route:", " -> ".join(route))
print("Minimum Distance:", min_cost)
```
